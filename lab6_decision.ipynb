```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor, plot_tree
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import accuracy_score, mean_squared_error
# 1. Load classification dataset (Iris)
url_classification = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/iris.csv"
columns_class = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width',
'class']
df_class = pd.read_csv(url_classification, names=columns_class)
df_class.head()
# Prepare classification data
X_class = df_class.drop('class', axis=1)
y_class = df_class['class']
Xc_train, Xc_test, yc_train, yc_test = train_test_split(X_class, y_class,
test_size=0.2, random_state=42)
# 3. Decision Tree Classifier
clf_tree = DecisionTreeClassifier()
clf_tree.fit(Xc_train, yc_train)
yc_pred_tree = clf_tree.predict(Xc_test)
print("Decision Tree Classifier Accuracy:", accuracy_score(yc_test,
yc_pred_tree))
# Plot the decision tree
plt.figure(figsize=(12, 6))
plot_tree(clf_tree, feature_names=X_class.columns, class_names=clf_tree.classes_,
filled=True)
plt.title("Decision Tree (Classification)")
plt.show()
# 4. Random Forest Classifier
clf_rf = RandomForestClassifier(n_estimators=100)
clf_rf.fit(Xc_train, yc_train)
yc_pred_rf = clf_rf.predict(Xc_test)
print("Random Forest Classifier Accuracy:", accuracy_score(yc_test, yc_pred_rf))
# 5. Load regression dataset (California housing)
url_regression = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/housing.csv"
columns_reg = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms', 'Population',
'AveOccup', 'Latitude', 'Longitude', 'MedHouseVal']
df_reg = pd.read_csv(url_regression, names=columns_reg)
# 6. Prepare regression data
X_reg = df_reg.drop('MedHouseVal', axis=1)
y_reg = df_reg['MedHouseVal']
Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_reg, y_reg,
test_size=0.2, random_state=42)
# 7. Decision Tree Regressor
reg_tree = DecisionTreeRegressor()
reg_tree.fit(Xr_train, yr_train)
yr_pred_tree = reg_tree.predict(Xr_test)
print("Decision Tree Regressor RMSE:", mean_squared_error(yr_test, yr_pred_tree,
squared=False))
# 8. Random Forest Regressor
reg_rf = RandomForestRegressor(n_estimators=100)
reg_rf.fit(Xr_train, yr_train)
yr_pred_rf = reg_rf.predict(Xr_test)
print("Random Forest Regressor RMSE:", mean_squared_error(yr_test, yr_pred_rf,
squared=False))
```
